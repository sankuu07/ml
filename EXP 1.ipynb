## 1. NumPy – Advanced Numerical Computing

```python

import numpy as np
np.random.seed(42)
```

```python

# Vectorized operations
A = np.random.randint(1, 10, (4, 4))
B = np.random.randint(1, 10, (4, 4))

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix Multiplication:\n", np.dot(A, B))
```

```python

# Eigen values and eigen vectors
eigen_values, eigen_vectors = np.linalg.eig(A)
print("Eigen Values:", eigen_values)
print("Eigen Vectors:\n", eigen_vectors)
```

## 2. Pandas – Realistic Data Analysis

```python

import pandas as pd
```

```python

# Simulated dataset
data = {
    'Student': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Math': [78, 85, 67, 90, 88, 76],
    'Physics': [72, 80, 65, 88, 91, 70],
    'Chemistry': [75, 82, 60, 85, 89, 74]
}

df = pd.DataFrame(data)
df
```

```python

# Feature engineering
df['Average'] = df[['Math', 'Physics', 'Chemistry']].mean(axis=1)
df['Result'] = np.where(df['Average'] >= 75, 'Pass', 'Fail')
df
```

```python

# Grouping and aggregation
df.groupby('Result').mean(numeric_only=True)
```

## 3. Matplotlib – Advanced Visualization

```python

import matplotlib.pyplot as plt
```

```python

# Histogram
plt.hist(df['Average'], bins=5)
plt.title("Distribution of Average Marks")
plt.xlabel("Average")
plt.ylabel("Frequency")
plt.show()
```

```python

# Box plot
plt.boxplot([df['Math'], df['Physics'], df['Chemistry']], labels=['Math','Physics','Chemistry'])
plt.title("Subject-wise Score Distribution")
plt.show()
```
